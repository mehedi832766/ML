Dummy_varriable


```python
df.rename({'Car Model':'Model'},axis=1,inplace=True)
df
```

```python
dummies=pd.get_dummies(df.Model)
dummies
```

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
final = merged.drop(['Model','Mercedez Benz C class'],axis=1)
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=final.drop(['Price'],axis=1)
X
```

```python
y=final['Price']
y
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

different approach


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
X1=df.drop(['Price'],axis=1)
X1
```

```python
ct = ColumnTransformer([("Model", OneHotEncoder(), [0])], remainder = 'passthrough')
X1=ct.fit_transform(X1)
X1
```

```python
X1=X1[:,1:]
X1
```

```python
model.fit(X1,y)
```

```python
model.predict([[0,0,45000,4]])
```
